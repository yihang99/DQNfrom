```python
import csv

with open('linear_double_dqn.csv','r') as csvfile:
    reader = csv.reader(csvfile, delimiter=',')
    avg_episode_rewards = [float(row[2]) for row in reader]
    print(avg_episode_rewards)

with open('linear_double_dqn.csv','r') as csvfile:
    reader = csv.reader(csvfile, delimiter=',')
    avg_truncated_episode_rewards = [float(row[1]) for row in reader]
    print(avg_truncated_episode_rewards)
```

```python
import matplotlib.pyplot as plt
import numpy as np

plt.ylabel('average truncated episode rewards')
plt.plot(avg_truncated_episode_rewards, label='avg_truncated_episode_rewards')
```

```python
import matplotlib.pyplot as plt
import numpy as np

plt.ylabel('average episode rewards')
plt.plot(avg_episode_rewards, label='avg_episode_rewards')
```
